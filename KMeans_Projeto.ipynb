```python
#vamos encontrar um k adequado
sse = {}
faixa = range(1,20)
for kclusters in faixa:
    kmeans1 = KMeans(n_clusters=kclusters)
    kmeans1.fit(df.drop('Private', axis=1))
    sse[kclusters] = kmeans1.inertia_ # Inertia: Soma das distâncias das amostras ao centro de aglomerado mais próximo

plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Número de cluster")
plt.ylabel("soma dos erros ao quadrado - SSE")
plt.title("Número de cluster por SSE - Algoritmo KMeans no SKLearn")
plt.show()
```

```python
def elbow(array):
    se = []
    for n in range(2, 21):
        kmeans = KMeans(n_clusters=n)
        kmeans.fit(X=array)
        se.append(kmeans.inertia_)

    x1, y1 = 2, se[0]
    x2, y2 = 20, se[len(se)-1]

    distances = []
    for i in range(len(se)):
        x0 = i+2
        y0 = se[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = math.sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distances.append(numerator/denominator)

    return distances.index(max(distances)) + 2
```

```python
elbow(normalized_df)
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []

faixa = range(2,20)

for n_cluster in faixa:
    silhouette_scores.append(
        silhouette_score(df.drop('Private', axis=1), KMeans(n_clusters = n_cluster).fit_predict(df.drop('Private', axis=1))))

# Plotando um gráfico de barras para comparar os resultados
k = list(faixa)
plt.bar(k, silhouette_scores)
plt.xlabel('Número de clusters', fontsize = 10)
plt.ylabel('Pontuação da silhueta', fontsize = 10)
plt.show()
```

```python
silhouette_scores.index(max(silhouette_scores))+2
```

```python
from sklearn.decomposition import PCA
```

```python
for i in range(0,16):
    pca = PCA(n_components=i)
    pca.fit(scaled_df)
    var_sum = pca.explained_variance_ratio_.sum()
    if var_sum >= 0.9:
        print(i)
        break
```
